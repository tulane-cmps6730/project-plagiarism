```python
# prompt: make the same function that cleans the text1 first

import re

def calculate_bleu_scores(text1, dataframe):
  bleu_scores = []
  for index, row in dataframe.iterrows():
    text1 = re.sub(r'[^\w\s]', '', text1)  # Remove punctuation
    text1 = text1.lower()  # Lowercase the text
    bleu_score = sentence_bleu([row['concatenated_text']], text1)
    bleu_scores.append(bleu_score)
  return bleu_scores

```
